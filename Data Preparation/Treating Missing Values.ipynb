```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
null = np.nan
series_obj = Series([3,67,null,4,6,null,8,3,null,2])
series_obj
```

```python
series_obj.isnull()
```

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
DF_obj.loc[3:5,0] = null
DF_obj.loc[1:4,5] = null
DF_obj
```

```python
## FIlling nulls with 0
filled_df = DF_obj.fillna(0)
filled_df
```

```python
## Filling nulls with desired values
filled_df = DF_obj.fillna({0: 0.1, 5: 1.24444})
filled_df
```

```python
## Filling nulls with previous non null number in the row.  ffill = fill forward method
fill_df = DF_obj.fillna(method='ffill')
fill_df
```

```python
## Counting the missing values in the row
DF_obj.isnull().sum()
```

```python
## Filtering out missing values row wise
df_no_null = DF_obj.dropna()
df_no_null
```

```python
## Filtering out missing values col wise
df_no_null = DF_obj.dropna(axis=1)
df_no_null
```
